```python
concrete_data_columns=concrete_data.columns
```

```python
predictors=concrete_data.drop(['Age','Strength'],axis='columns')
target=concrete_data['Strength']
```

```python
target.head()
```

```python
# Import Keras
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
n_cols=predictors.shape[1]  #no. of predictors
```

```python
n_cols
```

```python
#Building Neural network

def regression_model():
    #create model
    model=Sequential()
    model.add(Dense(10,activation='relu',input_shape=(n_cols,)))
    model.add(Dense(1))
    
    #compile the model
    model.compile(optimizer='adam',loss='mean_squared_error')
    return model
```

```python
#Spliting dataset into tarining and test data using train_test_split of sklearn

from sklearn.model_selection import train_test_split
```

```python
#Step 1
X_train,X_test,y_train,y_test=train_test_split(predictors,target,test_size=0.3,random_state=10)
#Step 2
model=regression_model()
model.fit(X_train,y_train,epochs=50)
```

```python
#Step 3
y_predicted=model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
ms1=mean_squared_error(y_test,y_predicted)
```

```python
ms1
```

```python
#Step 4

#Regression function using keras
def regression_keras(X,y,epo):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=10)
    model=regression_model()
    model.fit(X_train,y_train,epochs=epo)
    y_predicted=model.predict(X_test)
    mse=mean_squared_error(y_test,y_predicted)
    return mse
```

```python
#Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors.
#For loop for calculating the mean squared error list
global msq_error1
msq_error1=[]*50
epoch=50
for i in range(50):
    msq_error1.append(regression_keras(predictors,target,epoch))
```

```python
#Step 5
#Calculation of the mean and the standard deviation of the mean squared errors.
mean_msq=np.mean(msq_error1)
std_msq=np.std(msq_error1)
print("Mean is {}% and standard deviation is {}% for 50 epochs".format(mean_msq,std_msq))
```

```python
## PART 2 -With normalized data

#normalize the data by subtracting the mean and dividing by standard deviation
predictors_norm=(predictors-predictors.mean())/predictors.std()
predictors_norm.head()
```

```python
#Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors for normalized data

global msq_error2
msq_error2=[]*50
epoch=50
for i in range(50):
    msq_error2.append(regression_keras(predictors_norm,target,epoch))
```

```python
#Step 5
#Calculation of the mean and the standard deviation of the mean squared errors for normalized data.
mean_msq2=np.mean(msq_error2)
std_msq2=np.std(msq_error2)
print("Mean is {}% and standard deviation is {}% for 50 epochs for normalized data".format(mean_msq2,std_msq2))
```

```python
print("The difference between mean when normalized data used is {}".format(mean_msq2-mean_msq))
```

```python

def regression_model():
    #create model
    model=Sequential()
    model.add(Dense(10,activation='relu',input_shape=(n_cols,)))
    model.add(Dense(10,activation='relu'))                          #2nd hidden layer
    model.add(Dense(10,activation='relu'))                          #3rd hidden layer
    model.add(Dense(1))
    
    #compile the model
    model.compile(optimizer='adam',loss='mean_squared_error')
    return model
```

```python
global msq_error4
msq_error4=[]*50
epoch=50
for i in range(50):
    msq_error4.append(regression_keras(predictors_norm,target,epoch))
```

```python
#Calculation of the mean and the standard deviation of the mean squared errors for normalized data.
mean_msq4=np.mean(msq_error4)
std_msq4=np.std(msq_error4)
print("Mean is {}% and standard deviation is {}% for {} epochs for normalized data".format(mean_msq4,std_msq4,epoch))
```
